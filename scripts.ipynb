```python
import yfinance as yf
import pandas as pd
import math
```

```python
ticker_csv = pd.read_csv('ticker_subset.csv')
```

```python
stock_info = pd.DataFrame(columns=['Ticker', 'Stock_date', 'Open_Price', 'Close_Price', 'Volume'])

stock_info
```

```python
stocks_exchange = {'ADBE': 2, 'AMD': 2, 'AMZN': 2, 'AAPL': 2, 'BBY': 2, 'CVS': 2, 'GOOGL': 2, 'IBM': 1, 'TSLA': 2}
stocks_sector = {'ADBE': 8, 'AMD': 8, 'AMZN': 2, 'AAPL': 8, 'BBY': 2, 'CVS': 6, 'GOOGL': 1, 'IBM': 8, 'TSLA': 2}
corporation_id = {'ADBE': 11, 'AMD': 12, 'AMZN': 13, 'AAPL': 14, 'BBY': 15, 'CVS': 16, 'GOOGL': 17, 'IBM': 18, 'TSLA': 19}
```

```python
for ticker in ticker_csv['Tickers']:
    ticker_val = yf.Ticker(ticker)
    hist = ticker_val.history(period='1mo')
    for i in range(len(hist)):
        date = '{:04d}-{:02d}-{:02d}'.format(hist.index[i].year, hist.index[i].month, hist.index[i].day)
        new_row = {'Ticker':ticker, 'Stock_date': date, 
                    'Open_Price':round(hist.Open[i],2), 'Close_Price':round(hist.Close[i],2),
                    'Volume':hist.Volume[i], 'Exchange_ID': int(stocks_exchange[ticker]), 'Sector_ID': int(stocks_sector[ticker]),
                    'Corporation_ID': int(corporation_id[ticker])}
        print(new_row)
        stock_info = stock_info.append(new_row, ignore_index=True)
```

```python
stock_info.to_csv("stock_data.csv", index=False)
```
