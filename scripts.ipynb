```python
for ticker in ticker_csv['Tickers'][0:100]:
    ticker_val = yf.Ticker(ticker)
    hist = ticker_val.history(period='1y')
    for i in range(len(hist)):
        date = '{:04d}-{:02d}-{:02d}'.format(hist.index[i].year, hist.index[i].month, hist.index[i].day)
        new_row = {'Ticker':ticker, 'Date':date, 'Volume':hist.Volume[i], 
                    'Open_Price':round(hist.Open[i],2), 'Close_Price':round(hist.Close[i],2)}
        stock_info = stock_info.append(new_row, ignore_index=True)
        # print([ticker, date, hist.Volume[i], round(hist.Open[i],2), round(hist.Close[i],2)])
```
